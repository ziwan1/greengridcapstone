```python
import requests
import os
import shutil
```

```python
my_key = "a8337d2d1c46469a41c89a549367f123"
api_url = "http://api.positionstack.com/v1/forward"
MAIN_URL = f"{api_url}?access_key={my_key}&query="
```

```python
# request api untuk mendapatkan seluruh kabupaten di jawa timur
response_kab = requests.get("https://dev.farizdotid.com/api/daerahindonesia/kota?id_provinsi=35")
data = response_kab.json()
print(data)

# get id, and nama
for kab in data['kota_kabupaten']:
    print(kab['id'], kab['nama'])
```

```python
my_task = {}
# dapatkan nama kecamatan dari Magetan, Malang, Mojokerto, Nganjuk, Ngawi, dan Pacitan
for kab in data['kota_kabupaten']:
    # condition if nama contain word Magetan, Malang, Mojokerto, Nganjuk, Ngawi, or Pacitan
    if "Magetan" in kab['nama'] or "Malang" in kab['nama'] or "Mojokerto" in kab['nama'] or "Nganjuk" in kab['nama'] or "Ngawi" in kab['nama'] or "Pacitan" in kab['nama']:
        response_kec = requests.get(f"https://dev.farizdotid.com/api/daerahindonesia/kecamatan?id_kota={kab['id']}")
        data_kec = response_kec.json()
        for kec in data_kec['kecamatan']:
            # pisahkan tiap kecamatan tiap kabupaten dalam bentuk list
            my_task[kab['nama']] = [kec['nama'] for kec in data_kec['kecamatan']]

# print my_task beautifully
for key, value in my_task.items():
    print(key, value, len(value))

```

```python
# now request longitute latitue from every kecamatan in my_task, make a new dictionary with key is kecamatan name and value is longitute and latitude

my_task_lat_long = {}
for key, value in my_task.items():
    for kec in value:
        # pisahkan tiap kecamatan tiap kabupaten dalam bentuk list
        response_lat_long = requests.get(f"{MAIN_URL}{kec + ', East Java, Indonesia'}")
        data_lat_long = response_lat_long.json()
        my_task_lat_long[kec] = [data_lat_long['data'][0]['latitude'], data_lat_long['data'][0]['longitude']]
        print(kec, data_lat_long['data'][0]['latitude'], data_lat_long['data'][0]['longitude'])
        
```

```python
METEO_URL = "https://archive-api.open-meteo.com/v1/archive?"
ADDON = "&start_date=2010-02-01&end_date=2023-04-10&hourly=temperature_2m,relativehumidity_2m,dewpoint_2m,apparent_temperature,pressure_msl,surface_pressure,precipitation,rain,snowfall,weathercode,cloudcover,cloudcover_low,cloudcover_mid,cloudcover_high,shortwave_radiation,direct_radiation,diffuse_radiation,direct_normal_irradiance,windspeed_10m,windspeed_100m,winddirection_10m,winddirection_100m,windgusts_10m,et0_fao_evapotranspiration,vapor_pressure_deficit,soil_temperature_0_to_7cm,soil_temperature_7_to_28cm,soil_temperature_28_to_100cm,soil_temperature_100_to_255cm,soil_moisture_0_to_7cm,soil_moisture_7_to_28cm,soil_moisture_28_to_100cm,soil_moisture_100_to_255cm&timezone=Asia%2FSingapore&windspeed_unit=ms&timeformat=unixtime"
```

```python
for x, y in my_task_lat_long.items():
    n = 0
    # check if file exist iy tes continur, if not create file
    if os.path.isfile(f'{x}.csv'):
        print(f"{x} exist")
        continue
    else:
        response_meteo = requests.get(f"{METEO_URL}latitude={y[0]}&longitude={y[1]}{ADDON}")
        data_meteo = response_meteo.json()
        with open(f'{x}.csv', 'a') as f:
            # write hello world
            f.write("latitude,longitude,elevation,utc_offset_seconds,timezone,timezone_abbreviation\n")
            f.write(f"{data_meteo['latitude']},{data_meteo['longitude']},{data_meteo['elevation']},{data_meteo['utc_offset_seconds']},{data_meteo['timezone']},{data_meteo['timezone_abbreviation']}\n")
            f.write("\n")
            f.write("time,temperature_2m,relativehumidity_2m,dewpoint_2m,apparent_temperature,pressure_msl,surface_pressure,precipitation,rain,snowfall,weathercode,cloudcover,cloudcover_low,cloudcover_mid,cloudcover_high,shortwave_radiation,direct_radiation,diffuse_radiation,direct_normal_irradiance,windspeed_10m,windspeed_100m,winddirection_10m,winddirection_100m,windgusts_10m,et0_fao_evapotranspiration,vapor_pressure_deficit,soil_temperature_0_to_7cm,soil_temperature_7_to_28cm,soil_temperature_28_to_100cm,soil_temperature_100_to_255cm,soil_moisture_0_to_7cm,soil_moisture_7_to_28cm,soil_moisture_28_to_100cm,soil_moisture_100_to_255cm\n")
            for j in range(0, len(data_meteo['hourly']['time'])):
                f.write(f"{data_meteo['hourly']['time'][j]},{data_meteo['hourly']['temperature_2m'][j]}, {data_meteo['hourly']['relativehumidity_2m'][j]}, {data_meteo['hourly']['dewpoint_2m'][j]}, {data_meteo['hourly']['apparent_temperature'][j]}, {data_meteo['hourly']['pressure_msl'][j]}, {data_meteo['hourly']['surface_pressure'][j]}, {data_meteo['hourly']['precipitation'][j]}, {data_meteo['hourly']['rain'][j]}, {data_meteo['hourly']['snowfall'][j]}, {data_meteo['hourly']['weathercode'][j]}, {data_meteo['hourly']['cloudcover'][j]}, {data_meteo['hourly']['cloudcover_low'][j]}, {data_meteo['hourly']['cloudcover_mid'][j]}, {data_meteo['hourly']['cloudcover_high'][j]}, {data_meteo['hourly']['shortwave_radiation'][j]}, {data_meteo['hourly']['direct_radiation'][j]}, {data_meteo['hourly']['diffuse_radiation'][j]}, {data_meteo['hourly']['direct_normal_irradiance'][j]}, {data_meteo['hourly']['windspeed_10m'][j]}, {data_meteo['hourly']['windspeed_100m'][j]}, {data_meteo['hourly']['winddirection_10m'][j]}, {data_meteo['hourly']['winddirection_100m'][j]}, {data_meteo['hourly']['windgusts_10m'][j]}, {data_meteo['hourly']['et0_fao_evapotranspiration'][j]}, {data_meteo['hourly']['vapor_pressure_deficit'][j]}, {data_meteo['hourly']['soil_temperature_0_to_7cm'][j]}, {data_meteo['hourly']['soil_temperature_7_to_28cm'][j]}, {data_meteo['hourly']['soil_temperature_28_to_100cm'][j]}, {data_meteo['hourly']['soil_temperature_100_to_255cm'][j]}, {data_meteo['hourly']['soil_moisture_0_to_7cm'][j]}, {data_meteo['hourly']['soil_moisture_7_to_28cm'][j]}, {data_meteo['hourly']['soil_moisture_28_to_100cm'][j]}, {data_meteo['hourly']['soil_moisture_100_to_255cm'][j]}\n")
            print(f"{x} Done")
```
